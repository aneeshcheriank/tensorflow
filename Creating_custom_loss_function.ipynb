## Example
- Huber Loss
- [doc on huber loss](https://en.wikipedia.org/wiki/Huber_loss)

```python
def huber_loss(y_true, y_pred):
  threshold=1 # hardcoded threshold value (class for parameterizing the threshold)
  error = y_true - y_pred

  return tf.where(
      tf.abs(error) <= threshold,
      tf.square(error)/2,
      threshold*(tf.abs(error) - threshold/2)
  )  
```

- list functions to calculate the loss
  - the input value is a list

```python
model = tf.keras.Sequential([
    keras.layers.Dense(1, input_shape=[1])                             
])

model.compile(
    optimizer='sgd',
    loss = 'huber_loss'
)
```

```python
x = np.arange(-1, 4, 1, dtype=float)
y = 2*x + 3
```

```python
history = model.fit(
    x, y,
    epochs = 500
)
```
